```python
import pandas as pd

# Sample DataFrame
data = {
    'loan_amnt': [1000, 1500, 2000, 2500, 2500, 3000, 3000],
    'interest_rate': [10.5, 12.3, 15.2, 9.7, 9.7, 8.5, 8.5],
    'term': [36, 36, 36, 60, 60, 36, 36]
}
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Find the maximum value in the 'loan_amnt' column
max_value = df['loan_amnt'].max()

# Filter the DataFrame to get all rows with the maximum 'loan_amnt'
max_rows = df[df['loan_amnt'] == max_value]

# Display the rows with the maximum 'loan_amnt'
print("\nRows with the maximum loan amount:")
max_rows
```
